```python
#!/usr/bin/env python
# coding: utf-8

import glob
import os
from PIL import Image
import cv2
import numpy as np

#実行ディレクトリを取得
dir_BI2IC = os.getcwd()
#入力ディレクトリ名を指定
dir_input = dir_BI2IC + "\\input"
#出力ディレクトリ名を指定
dir_output = dir_BI2IC + "\\output"

print("Current Directory is : " + dir_BI2IC)
print("Input Directory is : " + dir_input)
print("Output Directory is : " + dir_output)

#取得するCSVファイルの名前規則設定ファイルを読み込み
fp_ftype_conf = open(dir_BI2IC + "\\configFileType.txt", "r")
file_type = ""
for line in fp_ftype_conf:
    file_type = line
fp_ftype_conf.close()
print(file_type)

#入力ディレクトリ内PGMファイルの名前リストを取得
bidx_name_list = glob.glob(dir_input + "\\" + file_type)

def convert_to_indexcolor(array, filename):
    # 虹色のカラーパレット
    color_palette = [
        255, 0, 0,
        255, 165, 0,
        255, 255, 0,
        0, 128, 0,
        0, 255, 255,
        0, 0, 255,
        128, 0, 128,
        255, 255, 255
    ]
    with Image.fromarray(array, mode='P') as img:
        img.putpalette(color_palette)  # パレットの設定
        img.save(dir_output + f'\\idxc_{filename}.png')
print(bidx_name_list)
for s in bidx_name_list:
    img = cv2.imread(s, 0)
    height, width = img.shape[:2]
    fname = s[-12:-4]
    for h in range(height):
        for w in range(width):
            if img[h, w] == 255:
                img[h, w] = 7
    convert_to_indexcolor(img, fname)
    #with Image.open(s) as img:
        #p_array = np.asarray(img)  # Numpy配列に変換
        #print(p_array.shape)
        #print(s)
        #convert_to_indexcolor(p_array, s)
print("complete saving!")
```
