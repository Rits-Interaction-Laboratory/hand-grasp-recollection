```python
import csv
import os
import matplotlib
#matplotlib.use('Agg')
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
sns.set_style("darkgrid")

PATH = os.getcwd()
target_epoch = 99
train_buffer = []
test_buffer = []
train_input_value = []
train_output_value = []
test_input_value = []
test_output_value = []

with open(PATH + "\\outputs\\network_outs_" + str(target_epoch) + ".csv", mode='r') as f:
    reader = csv.reader(f)
    test_flag = False
    for row in reader:
        if row:
            if row[0] == "test_test":
                test_flag = True
            if row and not(row[0].startswith("train")) and not(row[0].startswith("test")) and not(row[0].startswith("epoch:")) and not(row[0].startswith("batch:")):
                if len(row) != 3 and len(row) != 8:
                    if not test_flag:
                        train_buffer.append(row)
                    elif test_flag:
                        test_buffer.append(row)

print(len(train_buffer), len(test_buffer))
for b in range(int(len(train_buffer)/6)):
    input_x = []
    input_y = []
    input_z = []
    output_x = []
    output_y = []
    output_z = []
    for ix in train_buffer[b*6+0]:
        input_x.append(float(ix))
    for iy in train_buffer[b*6+1]:
        input_y.append(float(iy))
    for iz in train_buffer[b*6+2]:
        input_z.append(float(iz))
    for ox in train_buffer[b*6+3]:
        output_x.append(float(ox))
    for oy in train_buffer[b*6+4]:
        output_y.append(float(oy))
    for oz in train_buffer[b*6+5]:
        output_z.append(float(oz))
    train_input_value.append([input_x, input_y, input_z])
    train_output_value.append([output_x, output_y, output_z])
for b in range(int(len(test_buffer)/6)):
    input_x = []
    input_y = []
    input_z = []
    output_x = []
    output_y = []
    output_z = []
    for ix in test_buffer[b*6+0]:
        input_x.append(float(ix))
    for iy in test_buffer[b*6+1]:
        input_y.append(float(iy))
    for iz in test_buffer[b*6+2]:
        input_z.append(float(iz))
    for ox in test_buffer[b*6+3]:
        output_x.append(float(ox))
    for oy in test_buffer[b*6+4]:
        output_y.append(float(oy))
    for oz in test_buffer[b*6+5]:
        output_z.append(float(oz))
    test_input_value.append([input_x, input_y, input_z])
    test_output_value.append([output_x, output_y, output_z])    
```
