```python
a = torch.randn([5, 4])
b = torch.randn([5, 4])
mask = torch.tensor([[1.,0.,1.,1.],[1.,1.,1.,1.],[0.,0.,1.,1.],[1.,1.,1.,0.],[1.,1.,1.,1.]])
m_size = torch.tensor(mask.size()[1])
m_sum = torch.tensor(torch.sum(mask, 1))
print("m_size:",m_size)
print("m_sum:",m_sum)
rate = (m_size.item() / m_sum)
mse = nn.MSELoss()

masked_a = torch.mul(a, mask)
masked_b = torch.mul(b, mask)

result1 = mse(a, b)

#result2 = torch.mean(torch.pow(torch.sub(a, b), 2))
result2 = torch.sub(masked_a, masked_b)
print("sub2:",result2)
result2 = torch.pow(result2, 2)
print("pow2:",result2)
#result2 = torch.mean(result2)
#print("mean:",result2)
result2 = torch.sum(result2, 1)
print("sum2:",result2)
result2 = torch.mul(rate, result2)
print("gain2:",result2)
result2 = torch.mean(result2)
print("mean2:",result2)

result3 = torch.sub(a, b)
print("sub3:",result3)
result3 = torch.pow(result3, 2)
print("pow3:",result3)
#result2 = torch.mean(result2)
#print("mean:",result2)
result3 = torch.mean(result3, 1)
print("mean3:",result3)
result3 = torch.mean(result3)
print("mean3-2:",result3)

print("a:",a)
print("b:",b)
print("mask:",mask)
print("masked_a:",masked_a)
print("masked_b:",masked_b)

print("rate:",rate)
print("result1:",result1.item())
print("result2:",result2.item())
```
