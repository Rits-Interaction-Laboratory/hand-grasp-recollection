```python
import cv2
import os
import datetime
import random
import numpy as np
```

```python
PATH = "C:\\Users\\administrator\\CenterOfTheCircle"
IMG_SIZE = 128
MAX_RADIUS = 16
MIN_RADIUS = 6
RADIUS_RANDOM = False
GEN_IMG_NUM = 10000
```

```python
now = datetime.datetime.now()
strdt_now = '{year}{month:02}{day:02}_{hour:02}{minute:02}{second:02}'.format(year=now.year, month=now.month, day=now.day, hour=now.hour, minute=now.minute, second=now.second)
dataPath = PATH + "\\" + strdt_now
os.mkdir(dataPath)

for i in range(GEN_IMG_NUM):
    #半径の決定
    if RADIUS_RANDOM:
        radius = random.randint(MIN_RADIUS, MAX_RADIUS)
        print("radius = " + str(radius))
    else:
        radius = MAX_RADIUS
        print("radius = " + str(radius))
    
    #中心座標の決定
    rangeMin = 0 + radius
    rangeMax = IMG_SIZE - radius - 1
    centX = random.randint(rangeMin, rangeMax)
    centY = random.randint(rangeMin, rangeMax)
    
    #画像の生成
    img = np.zeros((IMG_SIZE, IMG_SIZE, 1), np.uint8)
    img[:,:] = 255
    cv2.circle(img, (centX, centY), radius, 0, thickness = 1)
    
    #画像名の決定
    now2 = datetime.datetime.now()
    name = '{year}{month:02}{day:02}_{hour:02}{minute:02}{second:02}'.format(year=now2.year, month=now2.month, day=now2.day, hour=now2.hour, minute=now2.minute, second=now2.second) + "_" + str(centX) + "_" + str(centY) + "_"
    print("name: " + name)
    imgPath = dataPath + "\\" + name + ".png"
    
    #画像の出力
    cv2.imwrite(imgPath, img)
```
