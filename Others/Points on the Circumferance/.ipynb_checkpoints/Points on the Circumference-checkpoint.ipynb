```python
import os
import random
import csv
import numpy as np
import matplotlib.pyplot as plt

path = os.getcwd()
data_dir = path + "\\dataset"
img_dir = path + "\\img"
id_txt = path + "\\id.txt"
os.makedirs(data_dir, exist_ok = True)
os.makedirs(img_dir, exist_ok = True)

radius = 30
origin_x = 0
origin_y = 0
num_pnt = 20
data_id = 0
data_num = 10000 #作成したいデータ数に書き換え

#data_id 読み込み
def read_id():
    with open(id_txt) as f:
        f_id = int(f.read())
    return f_id

#data_id 書き込み
def write_id(id_num = None):
    with open(id_txt, mode='w') as f:
        if id_num == None:
            id_num = "0"
        f.write(str(id_num))

#csv 出力
def write_csv(data, data_id):
    with open(data_dir + "\\{:0=6}.csv".format(data_id), 'w') as f:
        writer = csv.writer(f)
        writer.writerow(data)

#idファイルが存在していない場合は作成
if not os.path.isfile(id_txt):
    write_id()

for i in range(data_num):
    data_id = read_id()
    #print("data_id:", data_id)
    
    #グラフの描画先の準備
    fig = plt.figure()

    data = []
    xs = []
    ys = []
    org_deg = random.uniform(0, 3600) / 100
    for p in range(num_pnt):
        #print("point:", p)
        theta = np.deg2rad(org_deg + p * (360 / num_pnt))
        px = int(radius * np.cos(theta))
        py = int(radius * np.sin(theta))
        data.append(px)
        xs.append(px)
        data.append(py)
        ys.append(py)
    
    #生成したデータをプロット
    if i < 30:
        plt.scatter(xs, ys)
        fig.savefig(img_dir + "\\{:0=6}.png".format(data_id))
        plt.close()
    
    #print(len(data))
    write_csv(data, data_id)
    write_id(data_id + 1)
print("fin generating.")
```
